# **OPERATORS**

````python
a = 14
b = 44

print(a>b)
print(a<b)
print(a==b)
print(a!=b)
print(a>=b)
print(a<=b)
````

# **4. Bitwise Operator**

Bitiwise operators act on bits and perform bit by bit operations. These are used to operate on binary numbers.

1. Bitwise NOT

# **6. Identity Operator**

"is" and "isnot" are the identity operators are both used to check if two values are located on the same part of the memory. Two variables that are equal do not imply that they are indentical.

````python
a = 10
b = 20
c = a

print(a is not b)
print(a is c)
print(a is b and a is c)
````

# **7. Membership Operator**

"in" and "not in" are the membership operators that are used to test whether a value or variable is in a sequence.

````python
x = 24
y = 20
my_list = [10,20,30,40,50]

if (x not in my_list):
  print("x is not present in given list")
else:
  print("x is present in given list")

if (y in my_list):
  print("y is present in given list")
else:
  print("y is not present in given list")

````

# **CONDITIONAL STATEMENTS**

# **1. If Statement**

It is used to execute a block of code only when a specified condition evaluates to True.

````python
age = int(input("enter the age : "))
if age >= 18:
  print("Eligible to vote..")
````

# **Short Hand If**

It is used to write if statements in a single line. It is useful when only one statement needs to be execute.

````python
age = 19
if age>18 : print("Eligible to vote.")
````

# **2. If else Statement**

It is used to execute one block of code when the condition is True and another block when the condtion is False..


````python
age = 10
if age <= 12:
  print("Travel for free..")
else:
  print("Pay for ticket.")
````

# **3. If-elif-else Statement**

It is used to check multiple conditions in a program. It executes a block of code when its condition evaluates to True after previous conditions evaluate to False

````python
age = 25
if age <= 12:
  print("Child")
elif age <= 19:
  print("Teenager")
elif age <= 35:
  print("Young adult")
else:
  print("Adult")
````

# **4. Nested IF-else Statement**

It is an if-else statement placed inside another if or else block. It is used to check conditions within another condition

````python
age = 70
is_member = True

if age >= 60:
  if is_member:
    print("30% senior discount !")
  else:
    print("20% senior discount !")
else:
  print("Not eligible for a senior discount")
````

# **LOOPS**

# **1. for Loop**

A for loop iterates over a sequence(list, tuple, range etc.)

````python
for i  in range(1, 6):
  print(i)
````

Example 2: Iterate through a list

````python
fruits = ["Apple", "Banana", "Mango"]
for fruit in fruits:
  print(fruit)
````

# **2. while Loop**

A while loop runs as long as the condition is True.

````python
count = 1

while count <= 5:
  print(count)
  count += 1
````

# **3. Loop Control Statements**

A break stops the loop immediately

````python
for i in range(1, 10):
  if (i == 5):
    break
  print (i)
````

**B. Continue**

skips the current iteration

````python
for i in range(1, 6):
  if i == 3:
    continue
  print(i)
````

**C. Pass**

Placeholder Statement

````python
for i in range(5):
  pass

print("Loop Completed")
````

# **4. Nested Loops**

**A loop inside another loop**

````python
for i in range(1,4):
  for j in range( 1, 4):
    print(i, j)
````
